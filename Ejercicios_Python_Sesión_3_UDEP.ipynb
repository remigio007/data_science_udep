3) Crear las variables precio unitario = 1500 y número de productos = 2000, hallando la venta total. Elegir los tipos de variables correctos.

```python
unit_price = 1500
number_product = 2000
currency = "S/"
total_sale = unit_price * number_product
# Mi primera version print("La venta total de " + f"{number_product}" + " unidades a precio unitario de " + f"{unit_price}" + " es " + f"{total_sale:,}")
print(f"La venta total de {number_product} unidades con p.u. de {currency} {unit_price} es {currency} {total_sale:,}")
```

4) Hallar el promedio de las estaturas de tres personas que miden 1.65, 1.93 y 1.86 centímetros.

```python
heights = [1.65, 1.93, 1.86]
heights_avg = sum(heights)/len(heights)
# Imprimir los elementos de la lista heights_str = ", ".join(str(x) for x in heights)
print(f"El promedio de las siguientes estaturas {heights} es {heights_avg:.2f}")
```
