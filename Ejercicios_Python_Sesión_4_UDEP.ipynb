1) Elabore un programa en Python que imprima los elementos de una lista numérica que sean impares. Si no hay ningún elemento, que emita un mensaje que diga "No hay ningún elemento impar"

```python
import random as rd
print("|||Programa para determinar los números impares dentro de un rango aleatorio.\nSe debe ingresar un número(m) que será la cantidad de números aleatorios del rango que irá desde 1 hasta un número(n).|||\n")

# ---------------- Validaciones ----------------------
# Ingreso el número de veces que se ejecutara range(es decir la cantidad de números aleatorios)
while True:
  qty_range_input = input("Ingrese un número entero(m): ")
  try:
    qty_range = int(qty_range_input)
    break
  except ValueError:
    print("Número invalido! Ingrese un valor según lo especificado")

# Ingreso del número que será el tope del rango aleatorio
while True:
  stop_random_input = input("Ingrese un número entero(n): ")
  try:
    stop_random = int(stop_random_input)
    break
  except ValueError:
    print("Número invalido! Ingrese un valor según lo especificado")

#Lógica
number_list = list(rd.randint(1, stop_random) for _ in range(qty_range))
number_odd_list = []

for i in number_list:
  if i%2 != 0:
    number_odd_list.append(i)

if number_odd_list == 0:
  print("No existen números impares")
else:
  print(f"Los números impares son {number_odd_list} de la lista {number_list}")

```
